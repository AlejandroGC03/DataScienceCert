```python
#import librerias requeridas
import pandas as pd
import numpy as np
```

```python
# a. Extraer la información del archivo.
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/country_vaccinations.csv')
df.head()
```

```python
# b.1 Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas
df.info()
```

```python
# b.2 asegurar que las columnas con fechas sean del tipo datetime64.
df['date']=pd.to_datetime(df['date'])
df.info()
```

```python
# c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines,
#    en otras palabras, agrupar por vaccines y realizar sumatoria).
Vaccpermanufacturer=df[['total_vaccinations','vaccines']].groupby('vaccines').sum('total_vaccinations')
print(f'La cantidad de vacunas aplicadas por cada compania es:\n{Vaccpermanufacturer}')
```

```python
# d. Obtener la cantidad de vacunas aplicadas en todo el mundo.
totalworld=df['total_vaccinations'].sum()
print(f'Cantidad de vacunas aplicadas en todo el mundo: {totalworld:.0f}')
```

```python
# e. Calcular el promedio de vacunas aplicadas por país.
Meanpercountry=df[['total_vaccinations','country']].groupby('country').mean('total_vaccinations')
print(f'Promedio de vacunas aplicadas por pais: \n{Meanpercountry}')
```

```python
# f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.
VaccJan2021=df.loc[df['date']=='2021-01-29','total_vaccinations'].sum()
print(f'Cantidad de vacunas aplicadas el dia 29 de Enero del 2021: {VaccJan2021:.0f}')
```

```python
# g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias)
#    con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.
conDiferencias=df.copy()
conDiferencias['diferencias']=abs(conDiferencias['daily_vaccinations_raw']-conDiferencias['daily_vaccinations'])
conDiferencias['diferencias']
```

```python
# h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.
delta=df['date'].max()-df['date'].min()
print(f'Periodo de tiempo entre registro mas reciente y registro mas antiguo: {delta.days} dias')
```

```python
# i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac)
#    con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).
conCantidad=df.copy()
conCantidad['canVac']=conCantidad['vaccines'].str.split(',').str.len()
conCantidad['canVac']
```

```python
# j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.
antes20=df[df['date']<'12/20/2020']
antes20
```

```python
# k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.
pfizer=df[df['vaccines'].str.contains('Pfizer')]
pfizer
```

```python
#  l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer)
#     en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente
df['source_website'] = df['source_website'].astype(str)
with pd.ExcelWriter('/content/drive/MyDrive/Colab Notebooks/resultadosReto.xlsx',engine='xlsxwriter', engine_kwargs={'options': {'strings_to_urls': False}}, datetime_format='mm-dd-yyyy') as writer:
    conDiferencias.to_excel(writer, sheet_name="Diferencias")
    conCantidad.to_excel(writer, sheet_name="Cantidad")
    antes20.to_excel(writer, sheet_name="Antes2020")
    pfizer.to_excel(writer, sheet_name="Pfizer")
```
